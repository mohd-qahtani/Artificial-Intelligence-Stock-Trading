```python
#‘linear’ : Linear prediction in terms of the differenced endogenous variables.
#‘levels’ : Predict the levels of the original endogenous variables.


#What Is an Endogenous Variable?
#An endogenous variable is a variable in a statistical model that's changed or determined by its 
#relationship with other variables within the model.
#In other words, an endogenous variable is synonymous with a dependent variable, 
#meaning it correlates with other factors within the system being studied. 
#Therefore, its values may be determined by other variables.

arima_pred = arima_result.predict(start = len(train_data), end = len(df)-1, typ="levels").rename("ARIMA Predictions")
arima_pred.head(20)
```

```python
scaled_train_data = scaler.transform(train_data)
```

#Before creating LSTM model we should create a Time Series Generator object.

```python
from keras.preprocessing.sequence import TimeseriesGenerator

n_input = 5
n_features= 9
generator = TimeseriesGenerator(scaled_train_data, train_data['Close'], length=n_input, batch_size=50)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM
```

```python


lstm_model = Sequential()
lstm_model.add(LSTM(200, input_shape=(n_input, n_features)))
lstm_model.add(Dense(1,activation='relu'))
lstm_model.add(Dropout(0.10))
lstm_model.compile(optimizer='adam', loss='mse')

lstm_model.summary()
```

```python
lstm_model.fit_generator(generator,epochs=50)
```

```python
lstm_model.get_config()
```

```python
import matplotlib.pyplot as plt
```

```python
losses_lstm = lstm_model.history.history['loss']
plt.figure(figsize=(12,4))
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.plot(range(len(losses_lstm)),losses_lstm);
```
